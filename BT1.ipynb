Rời rạc hoá dữ liệu trường Age,Income,Credit_rating với Binning

```python
def binning(col,cut_points,labels=None):
    minval=col.min()
    maxval=col.max()
    breaks_points=[minval]+cut_points+[maxval]
    colBin=pd.cut(col,bins=breaks_points,labels=labels,include_lowest=True)
    return colBin
    
```

```python
cutpoints_Age=[30,40]
df['Age']=binning(df['Age'],cutpoints_Age,['khoang 20-30','khoang 30-40','40 tro len'])
```

```python
df
```

```python
cutpoints_Income=[29999,59999]
df['Income']=binning(df['Income'],cutpoints_Income,['<30000','30000->60000','>60000'])
```

```python
df
```

```python
cutpoints_Credit=[3]
df['Credit Rating']=binning(df['Credit Rating'],cutpoints_Credit,['<4','>=4'])
```
